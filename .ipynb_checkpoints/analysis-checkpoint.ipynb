```python
import seaborn as sns
import matplotlib as plt
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importamos nuestra tabla limpia

df_data_2 = pd.read_csv('attacks_clean.csv')
df_data_2
```

### 1. PIE CHART
- Como pirmera grafica para defender mi análisis de datos sobre los ataques de los tiburones he elegido la gráfica pie chart.
- Al jugar con dos variables categóricas en el momento de expresar los datos mediante gráficas es mucho más complicado porque el tipo de gráficas a utilizar se limitan.
- Estas dos graficas (Pie chart de la variable 'Activity' y Pie chart de la variable 'Country') representan de manera muy visual y muy sencilla los porcetajes de los mayores ataques de tiburon. Como podemos observar:
   - Surfing --> con un 29.66 %
   - USA --> con un 36.55%

```python
plot = df_data_2['Activity'].value_counts().plot(kind='pie', autopct='%.2f', 
                                            figsize=(20, 20),
                                            title='Activity')
```

```python
plot = df_data_2['Country'].value_counts().plot(kind='pie', autopct='%.2f', 
                                            figsize=(20, 20),
                                            title='Country')
```

# 2. FACED GRID
- La segunda gráfica que he escogido para defender este análisis ha sido: Faced Grid.

- Como podemos observar en la gráfica, la hipótesis por la que he empezado este trabajo no era del todo cierta. El número de ataques por tiburón se encuentran con gran diferencia en aquellas personas que hacen surf pero NO en Autralia.
